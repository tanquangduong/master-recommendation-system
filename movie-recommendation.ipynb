# Recommendation engine

```python
import pandas as pd 
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
warnings.filterwarnings('ignore')
```

# Data exploration

```python
df_reviews = pd.read_csv('./datasets/reviews.csv')
df_reviews.head()
```

```python
df_reviews.describe()
```

```python
df_reviews['userId'].value_counts()
```

# Import Movie file dataset

```python
df_movie_titles = pd.read_csv('./datasets/movies.csv',index_col=False)
df_movie_titles.head()
```

# Merge the two dataset

```python
df = pd.merge(df_reviews, df_movie_titles, on='movieId')
df.head()
```

```python
df.describe()
```

# Use aggregation to see the average rating and number of ratings for each movie title

```python
df_ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
df_ratings['number_of_ratings'] = df.groupby('title')['rating'].count()
df_ratings.head()
```

# Convert the data from narrow to wide format

```python
movie_matrix = df.pivot_table(index='userId', columns='title', values='rating')
movie_matrix.head()
```

```python
movie_matrix.shape
```

# Recommender movies if user has watched Avatar (2009) and Thor(2011)

```python
Avatar_user_rating = movie_matrix['Avatar (2009)']
Avatar_user_rating = Avatar_user_rating.dropna()
Avatar_user_rating.head()
```

# Calculate correlation based on rating and number of rating

```python
similar_to_Avatar=movie_matrix.corrwith(Avatar_user_rating)
corr_Avatar = pd.DataFrame(similar_to_Avatar, columns=['correlation'])
corr_Avatar.dropna(inplace=True)
corr_Avatar = corr_Avatar.join(df_ratings['number_of_ratings'])
corr_Avatar.head()
```

# Set a threshold 100 for number of rating

```python
corr_Avatar[df_ratings['number_of_ratings'] > 100].sort_values(by='correlation', ascending=False).head(10)
```
